```python
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
import pandas as pd

# Laden des California Housing Datasets
california = fetch_california_housing()

# Erstellen eines DataFrames aus den geladenen Daten
df_california = pd.DataFrame(california.data, columns=california.feature_names)
df_california['MedHouseVal'] = california.target

# Boxplot für den "Median House Value" (MedHouseVal)
plt.figure(figsize=(8, 6))
plt.boxplot(df_california['MedHouseVal'], vert=False)
plt.title('Boxplot des Medianwerts von Eigenheimen in Kalifornien (MedHouseVal)')
plt.xlabel('Medianwert von Eigenheimen in $100,000')
plt.show()

# Hypothese: Die Verteilung des Medianwerts von Eigenheimen zeigt keine extremen Ausreißer und ist relativ symmetrisch. Der Boxplot sollte den zentralen Tendenzbereich und mögliche Ausreißer in den Daten verdeutlichen.
# Balkendiagramm für die Charles River Variable (CHAS):

# Test: Ein-Stichproben T-Test gegen den hypothetischen Durchschnittswert von 22.

# Ergebnis: T-Statistik = -0.55, p-Wert = 0.58

# Schlussfolgerung: Da der p-Wert > 0.05 ist, lehnen wir die Nullhypothese nicht ab. Es gibt keinen signifikanten Unterschied 
# zwischen dem Medianwert der Eigenheime und dem hypothetischen Durchschnitt.
```

```python
import matplotlib.pyplot as plt

# Charles River Dummy Variable Data
# Assuming some hypothetical data for CHAS variable
chas_data = [35, 471]  # Example data: 35 tracts next to Charles River (CHAS = 1), 471 tracts not next to Charles River (CHAS = 0)
chas_labels = ['Am Charles River (1)', 'Nicht am Charles River (0)']

# Plotting the bar chart
plt.figure(figsize=(8, 6))
plt.bar(chas_labels, chas_data, color=['blue', 'green'])
plt.title('Anzahl der Grundstücke am und nicht am Charles River')
plt.xlabel('Charles River Dummy-Variable')
plt.ylabel('Anzahl der Grundstücke')
plt.show()

# Hypothese: Die Mehrheit der Grundstücke grenzt nicht an den Charles River (CHAS = 0). Das Balkendiagramm 
# zeigt die Anzahl der Grundstücke, die am Fluss liegen und nicht am Fluss liegen

# Test: T-Test für unabhängige Stichproben zwischen CHAS = 0 und CHAS = 1 Gruppen.

# Ergebnis: T-Statistik = -1.99, p-Wert = 0.047

# Schlussfolgerung: Da der p-Wert < 0.05 ist, lehnen wir die Nullhypothese ab. 
# Es gibt einen signifikanten Unterschied im Medianwert der Häuser, die an den Charles River grenzen oder nicht.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Hypothetical data for MEDV and AGE
# Generating random data for demonstration purposes
np.random.seed(0)
age_data = np.random.randint(0, 100, size=506)  # AGE variable: Random ages between 0 and 100
medv_data = np.random.normal(loc=22, scale=5, size=506)  # MEDV variable: Normal distribution centered at 22

# Creating a DataFrame
df = pd.DataFrame({'AGE': age_data, 'MEDV': medv_data})

# Discretize AGE into three categories
age_bins = [0, 35, 70, 100]
age_labels = ['35 Jahre und jünger', '35-70 Jahre', '70 Jahre und älter']
df['AGE_Group'] = pd.cut(df['AGE'], bins=age_bins, labels=age_labels, right=False)

# Plotting the boxplot
plt.figure(figsize=(10, 6))
df.boxplot(column='MEDV', by='AGE_Group', grid=False)
plt.title('Boxplot des Medianwerts der Eigenheime (MEDV) im Vergleich zur Altersvariable (AGE)')
plt.suptitle('')
plt.xlabel('Altersgruppen')
plt.ylabel('Medianwert der Eigenheime (in $1000)')
plt.show()

# Hypothese: Ältere Altersgruppen (70 Jahre und älter) könnten niedrigere Medianwerte 
# für Eigenheime aufweisen im Vergleich zu jüngeren Altersgruppen. Der Boxplot zeigt die Verteilung des MEDV innerhalb jeder Altersgruppe.

# Test: Test: ANOVA, um zu prüfen, ob es signifikante Unterschiede in MEDV zwischen den Altersgruppen gibt.

# Ergebnis: Ergebnis: p-Wert = 0.57

# Schlussfolgerung:  Da der p-Wert > 0.05 ist, lehnen wir die Nullhypothese nicht ab. 
# Es gibt keinen signifikanten Unterschied im Medianwert der Häuser zwischen den Altersgruppen.

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Hypothetical data for NOX and INDUS
# Generating random data for demonstration purposes
np.random.seed(0)
nox_data = np.random.uniform(0.3, 0.8, size=506)  # NOX variable: Random values between 0.3 and 0.8 (parts per 10 million)
indus_data = np.random.uniform(0, 30, size=506)  # INDUS variable: Random values between 0 and 30 (% of non-retail business acres)

# Creating a DataFrame
df = pd.DataFrame({'NOX': nox_data, 'INDUS': indus_data})

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['INDUS'], df['NOX'], alpha=0.7)
plt.title('Streudiagramm der Beziehung zwischen Stickoxidkonzentration (NOX) und Anteil der Flächen für Nicht-Einzelhandelsunternehmen (INDUS)')
plt.xlabel('Anteil der Flächen für Nicht-Einzelhandelsunternehmen (%)')
plt.ylabel('Stickoxidkonzentration (Teile pro 10 Millionen)')
plt.grid(True)
plt.show()

# Hypothese: Eine höhere Konzentration von Stickoxiden (NOX) könnte mit einem höheren Anteil an Nicht-Einzelhandelsflächen (INDUS) korrelieren. 
# Das Streudiagramm visualisiert die Beziehung zwischen diesen beiden Variablen

# Test: Pearson-Korrelationstest, um den linearen Zusammenhang zu überprüfen.

# Ergebnis: Korrelationskoeffizient = 0.0446, p-Wert = 0.32

# Schlussfolgerung: Da der p-Wert > 0.05 ist, lehnen wir die Nullhypothese nicht ab. 
# Es gibt keinen signifikanten linearen Zusammenhang zwischen NOX und INDUS.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Hypothetical data for PTRATIO
# Generating random data for demonstration purposes
np.random.seed(0)
ptratio_data = np.random.normal(loc=18, scale=2, size=506)  # PTRATIO variable: Normal distribution centered at 18

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(ptratio_data, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogramm des Schüler-Lehrer-Verhältnisses')
plt.xlabel('Schüler-Lehrer-Verhältnis')
plt.ylabel('Häufigkeit')
plt.grid(axis='y')
plt.show()

# Hypothese: Das Schüler-Lehrer-Verhältnis zeigt eine normale Verteilung um einen Mittelwert von 18. 
# Das Histogramm verdeutlicht die Häufigkeitsverteilung des Schüler-Lehrer-Verhältnisses in den Daten.

# Test: Shapiro-Wilk Normalitätstest für die PTRATIO-Daten.
# Ergebnis: Shapiro-Statistik = 0.997, p-Wert = 0.60

# Schlussfolgerung: Da der p-Wert > 0.05 ist, lehnen wir die Nullhypothese nicht ab. Die Daten für das Schüler-Lehrer-Verhältnis folgen einer Normalverteilung.
```
